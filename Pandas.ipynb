# Pandas Dataframe
# Creating Dataframe from various method

```python
# Creating a DataFrame from a Dictionary

data = {'Name' : ['Alice', 'Bob', 'Charlie'],
        'Age' : [25, 30, 22],
        'Salary' : [50000, 60000, 45000]
       }

df_dict = pd.DataFrame(data)
df_dict
```

```python
#Creating a DataFrame from Lists

data_list = [['Alice', 25, 50000], ['Bob', 30, 60000], ['Charlie', 22, 45000]]
columns = ['Name', 'Age', 'Salary']

df_list = pd.DataFrame(data_list, columns = columns)
print(df_list)
```

```python
#Creating a Dataframe from Numpy Array

import numpy as np

data_array = np.array([['Alice', 25, 50000], ['Bob', 30, 60000], ['Charlie', 22, 45000]])

df_array = pd.DataFrame(data_array, columns = columns)
print(df_array)
```

```python
#Creating a DataFrame from a CSV file

df_csv = pd.read_csv('HousePrices.csv')
print(df_csv)
```

```python
#Creating a DataFrame from Excel File

df_csv = pd.read_excel('Iris.xlsx')
df_csv
```

# Working with Sample DataFrame

```python
data = {'Column name' : [5,15,8],
        'Column 1' : [10,20,30],
        'Column 2' : [100,200,300],
        'Another_column' : [25,35,45]
       }
df = pd.DataFrame(data)
df
```

```python
#Accessing a single column

column_data = df['Column name']
print('Single Column: ')
print(column_data)
```

```python
#Accessing multiple column

selected_column = df[['Column 1', 'Column 2']]
print("\nMultiple Columns")
print(selected_column)
```

```python
#Accessing a specific row by index

row_data = df.iloc[0]
print("Specific row: ")
print(row_data)
```

```python
#Accessing rows based on conditons 

filtered_rows = df[df['Column name'] > 10]
print("filtered rows")
print(filtered_rows)
```

```python
#Accessing a single cell by label

value = df.at[0, "Column name"]
print("Single cell by label: ")
print(value)
```

```python
#Accessing a single cell by position

value = df.iat[0,1]
print("single cell by position")
print(value)
```

```python
#Accessing data using loc

Selected_data = df.loc[0, 'Column name']
print("data using loc:")
print(Selected_data)
```

```python
#Conditional Access

Selected_Data = df[df['Column name'] > 10]
print("Conditional Access")
print(Selected_Data)
```
